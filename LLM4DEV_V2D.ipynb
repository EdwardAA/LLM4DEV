````python
#Returns a model-generated response for a single user prompt using the specified LLM.
from openai import OpenAI

client = OpenAI(api_key=OPENAI_API_KEY)

def get_completion(prompt, model="gpt-5.1"):
    completion = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0  # controls randomness
    )
    # NOTE: use `.content` here, not ["content"]
    return completion.choices[0].message.content
````

# **Section 1: Prompting Principles**
## **1.   List item: Writing Clear and Specific Instructions**







**Tactic 1: Use delimiters to clearly indicate distinct parts of the input**



Delimiters can be anything like: ```, """, < >, <tag> </tag>, :

````python
text = """
A machining process requires selecting proper cutting
parametersâ€”such as spindle speed, feed rate, and depth
of cutâ€”to achieve desired surface finish and dimensional
accuracy. Choosing these parameters involves balancing
tool wear, material removal rate, and thermal effects
on the workpiece. Providing clear specifications and
constraints helps ensure consistent part quality and
reduces costly rework. Detailed instructions allow
manufacturing engineers to optimize the process for
productivity and reliability.
"""

prompt = f"""
Summarize the following text into one sentence:

```{text}```
"""

response = get_completion(prompt)
print(response)
````

````python
import warnings
from bokeh.util.warnings import BokehUserWarning

warnings.filterwarnings("ignore", category=BokehUserWarning)
````

````python
import panel as pn
pn.extension()

# ---------------------------------------
# 1. CREATE CHAT PANEL FIRST
# ---------------------------------------
conversation_panel = pn.Column()   # persistent chat window

# ---------------------------------------
# 2. CREATE INPUT WIDGETS
# ---------------------------------------
inp = pn.widgets.TextInput(value="Hi", placeholder="Enter text hereâ€¦")
button_conversation = pn.widgets.Button(name="Chat!")

# ---------------------------------------
# 3. SYSTEM CONTEXT (MANUBOT)
# ---------------------------------------
context = [{
    "role": "system",
    "content": """
You are ManuBot, an automated assistant that helps engineers
specify and track manufacturing orders for machined and
sheet-metal parts. You greet the user, gather details
(material, process, tolerances, finish, quantity), and
summarize the order. Respond clearly and professionally.
"""
}]

# ---------------------------------------
# 4. OPENAI FUNCTION
# ---------------------------------------
def get_completion_from_messages(messages, model="gpt-4.1", temperature=0):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature)
    return response.choices[0].message.content

# ---------------------------------------
# 5. CALLBACK FUNCTION
# ---------------------------------------
def collect_messages(event):
    prompt = inp.value
    inp.value = ""  # clear input box

    # Add user message
    context.append({"role": "user", "content": prompt})
    conversation_panel.append(
        pn.Row("User:", pn.pane.Markdown(prompt, width=600))
    )

    # Get response
    response = get_completion_from_messages(context)
    context.append({"role": "assistant", "content": response})

    # Add assistant message
    conversation_panel.append(
        pn.Row("Assistant:", pn.pane.Markdown(response, width=600))
    )

button_conversation.on_click(collect_messages)

# ---------------------------------------
# 6. BUILD DASHBOARD (NOW EVERYTHING IS DEFINED)
# ---------------------------------------
dashboard = pn.Column(
    inp,
    pn.Row(button_conversation),
    conversation_panel
)

dashboard
````
